```python
from numpy import random
from matplotlib import pyplot as plt
```

# Ex 1: Points from 2D normal distribution

```python
mean = [-5, 5]
cov = [[1, 0], [0, 1]]

points = random.multivariate_normal(mean, cov, 200)

plt.plot(points[:, 0], points[:, 1], '.')
```

# Ex 2: 2 Clusters with different mean and covariances

```python
mean1 = [2, 2]
cov1 = [[3, 0], [0, 3]]
points1 = random.multivariate_normal(mean1, cov1, 200)

mean2 = [-1, -1]
cov2 = [[2, 0], [0, 2]]
points2 = random.multivariate_normal(mean2, cov2, 200)

plt.plot(points1[:, 0], points1[:, 1], '.g')
plt.plot(points2[:, 0], points2[:, 1], '.b')
```
